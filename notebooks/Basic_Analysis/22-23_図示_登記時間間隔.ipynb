# 複数の登記原因で以下を出す（ホークス過程に当てれるかを確かめる）
    - その登記原因だけを抽出
    - その登記原因での回数の集計（全体）
    - 物件ごとに回数を集計
    - 回数のヒストグラムを出す
    - Maxの回数である物件で図示

```python
df_buy = pd.read_csv("C:\後藤研\修士研究\data\登記受付帳22-23_所有権移転売買.csv")
```

```python
df_buy.head()
```

```python
len(df_buy)
```

```python
df_buy_land = df_buy[df_buy["type"]=='土地']
```

```python
df_buy_land.head()
```

```python
len(df_buy_land.groupby(['pref', 'location', 'land_num']))
```

```python
# 「pref」「location」「land_num」でグルーピングして登記回数を集計
grouped_data = df_buy_land.groupby(['pref', 'location', 'land_num']).size().reset_index(name='count')
```

```python
len(grouped_data)
```

```python
grouped_obj = df_buy_land.groupby(['pref', 'location', 'land_num'])
print("Group count from groupby:", len(grouped_obj))
print("Group count from grouped_data:", len(grouped_data))
```

```python
# 欠損値を確認
print(df_buy_land[['pref', 'location', 'land_num']].isna().sum())

# グループのキーを確認
print(grouped_obj.groups.keys())
print(grouped_data[['pref', 'location', 'land_num']].head())
```
