```python
import re
from fractions import Fraction

# 「２分の１」のような文字列を「1/2」のような数値に変換する関数
def convert_fraction(string):
    match = re.match(r'(\d+)分の(\d+)', str(string))  # str()を使用して文字列に変換
    if match:
        numerator = int(match.group(1))
        denominator = int(match.group(2))
        result = Fraction(numerator, denominator)
        return str(result)
    else:
        return str(string)

# df_2のownership_share列に対して関数を適用
df_2['ownership_share'] = df_2['ownership_share'].apply(convert_fraction)
```

```python
# df_1での各都道府県の出現割合を計算し、表にまとめる
df_1_counts = df_1['都道府県'].value_counts(normalize=True).reset_index()
df_1_counts.columns = ['都道府県', 'df_1_PofO']
df_1_counts['df_1_PofO'] *= 100  # パーセンテージに変換
df_1_counts['df_1_PofO'] = df_1_counts['df_1_PofO'].round(2)  # 有効数字4桁まで表示

# df_2での各都道府県の出現割合を計算し、表にまとめる
df_2_counts = df_2['pre'].value_counts(normalize=True).reset_index()
df_2_counts.columns = ['都道府県', 'df_2_PofO']
df_2_counts['df_2_PofO'] *= 100  # パーセンテージに変換
df_2_counts['df_2_PofO'] = df_2_counts['df_2_PofO'].round(2)  # 有効数字4桁まで表示

# df_1_countsとdf_2_countsを都道府県をキーにして結合
result = pd.merge(df_1_counts, df_2_counts, on='都道府県', how='outer')

# 「出現割合_df_1-出現割合_df_2」の列を追加
result['(df_1 - df_2)_PofO'] = result['df_1_PofO'] - result['df_2_PofO']

# 背景色を設定する関数
def highlight_cells(series):
    return ['background-color: orange' if abs(val) >= 7 else '' for val in series]

# 「出現割合_df_1-出現割合_df_2」の列にスタイルを適用
styled_result = result.style.apply(highlight_cells, subset=['(df_1 - df_2)_PofO'])

# スタイル付きのDataFrameをExcelファイルとして保存
styled_result.to_excel('pre_PofO.xlsx', index=False)

# 保存完了メッセージ
print('The styled result has been saved to pre_PofO.xlsx.')
```
